## Creación de nuevas columnas

Para la creación de nuevas columnas en un data frame puede pensarse en este como un diccionario. Esto debido a que como sabemos para crear una nueva llave dentro de un diccionario simplemente referenciamos la nueva llave y le asignamos un valor. Sin embargo, es una buena práctica definir el tipo de valor que estamos guardando con la función <code>.astype()</code>, la cual recibe como parámetro el tipo. e.g. <code>int, float, str, list, dict, etc.</code>.

```python
# Nuevas columnas: Minutos por partido y Goles por partido

epl_df['MinsPerMatch'] = (epl_df['Mins'] / epl_df['Matches']).astype(int)

epl_df['GoalsPerMatch'] = (epl_df['Goals'] / epl_df['Matches']).astype(float)

epl_df.iloc[:, [-2,-1]]
```

## Cálculos

Realizamos unos cuantos cálculos de información de interés, como lo son los goles y los penales.

```python
# Total de goles anotados
Total_goals = epl_df['Goals'].sum()
print(Total_goals)

# Total de goles anotados de penal
Total_penaltygoals = epl_df['Penalty_Goals'].sum()
print(Total_penaltygoals)
```

```python
# Penales intentados
Total_penaltyattempts = epl_df['Penalty_Attempted'].sum()
print(Total_penaltyattempts)

# Proporción
proportion_scored_v_attempted = Total_penaltygoals / Total_penaltyattempts
print(proportion_scored_v_attempted)
```

## Gráficos

Para realizar un gráfico de torta usaremos la librería <code>matplotlib (plt)</code>.
<br>
<ol>
    <li>
        Creamos la figura <code>plt.figure()</code>. Aquí podemos modificar el tamaño <code>figsize</code> en pulgadas, el color de fondo (background) <code>facecolor</code>, el color del borde (edge) <code>edgecolor</code>.
        <br>
        <a href='https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.figure.html'>Documentation matplotlib figure</a>
    </li>
    <br>
    <li>
        Creamos el tipo de gráfico que pondremos en la figura, en este caso es un pie chart por lo que usamos <code>plt.pie()</code>. La cual recibe distintos parámetros: 
        <ul>
            <li>
                <b>x:</b> Aquí ponemos nuestros datos, es un array de 1 dimensión.
            </li>
            <li>
                <b>labels:</b> Una lista con los nombres de cada dato.
            </li>
            <li>
                <b>colors:</b> Seleccionamos los colores que queremos usar. En este caso, por la librería seaborn usamos un set de colores pre-definido.
            </li>
            <li>
                <b>autopct:</b> el formato en el cual mostramos los valores de la gráfica. Si lo mostramos como porcentaje del total, si mostramos el número de datos es cada categoría o ambos.
            </li>
        </ul>
        <br>
        <a href='https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html'>Documentation matplotlib pie</a>
    </li>
    <br>
    <li>
        Usamos <code>plt.show()</code> para mostrar o imprimir la figura
    </li>
</ol>

```python
# Diagrama de torta para los penales

plt.figure(figsize=(6,6), 
           facecolor='#FFFFFF', 
           edgecolor='#000000')

penal_not_scored = epl_df['Penalty_Attempted'].sum() - epl_df['Penalty_Goals'].sum()

data = [penal_not_scored, Total_penaltygoals]
labels = ['Penalties missed', 'Penalties scored']
color = sns.color_palette('Set2')

plt.pie(data,
        labels=labels,
        colors=color,
        autopct='%.2f%%')

plt.show()
```
